# Value-at-Risk (VaR) Modeling and Backtesting

This project estimates potential portfolio losses using Value at Risk (VaR) and evaluates model performance using backtesting and Monte Carlo simulation.

## 1. Data Collection

```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv"

data = pd.read_csv(url)

# use closing price
prices = data["AAPL.Close"]

print(prices.head())
```

## 2. Returns Calculation

Daily returns are calculated to measure percentage change in price.

```python
# convert to returns
returns = prices.pct_change().dropna()

portfolio_returns = returns

print(portfolio_returns.head())
```

## 3. Historical VaR

Historical VaR estimates potential loss using past return distribution.

```python
# historical VaR
import numpy as np

# 95% VaR
var_95 = np.percentile(portfolio_returns, 5)

# 99% VaR
var_99 = np.percentile(portfolio_returns, 1)

print("95% VaR:", var_95)
print("99% VaR:", var_99)
```

## 4. Backtesting

Backtesting evaluates how often actual losses exceed predicted VaR.

```python
# Backtesting
# count violations
violations_95 = (portfolio_returns < var_95).sum()
violations_99 = (portfolio_returns < var_99).sum()

total_days = len(portfolio_returns)

print("95% violations:", violations_95)
print("99% violations:", violations_99)
print("Total days:", total_days)
```

```python
# convert to percentage
violation_rate_95 = violations_95 / total_days
violation_rate_99 = violations_99 / total_days

print("95% violation rate:", violation_rate_95)
print("99% violation rate:", violation_rate_99)
```

## 5. Monte Carlo Simulation

Monte Carlo simulation generates possible future returns assuming normal distribution.

```python
# Monte Carlo VaR
# parameters from data
mean = portfolio_returns.mean()
std = portfolio_returns.std()

# simulate returns
simulations = np.random.normal(mean, std, 10000)

# calculate VaR
mc_var_95 = np.percentile(simulations, 5)
mc_var_99 = np.percentile(simulations, 1)

print("Monte Carlo 95% VaR:", mc_var_95)
print("Monte Carlo 99% VaR:", mc_var_99)
```

## 6. Key Insights

- Historical VaR captures real extreme market movements.
- Monte Carlo VaR underestimates tail risk due to normal distribution assumption.
- Backtesting shows violation rates close to expected levels, indicating reasonable model calibration.
